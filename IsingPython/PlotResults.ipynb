```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

%matplotlib inline
```

```python
fig, ax = plt.subplots(1,1,figsize=(10,5))

Ts = np.array([2, 6, 8, 9, 10, 11, 12, 15, 20, 30, 40])/10 #x axis

for size in [11, 21, 51, 101, 121, 151]:
    Cs = []
    for T in [2, 6, 8, 9, 10, 11, 12, 15, 20, 30, 40]:
        temp_result = np.loadtxt('Results/T{}N{}E.txt'.format(T,size))[10000:] # tail -90000
        E = np.sum(temp_result)
        E2 = np.sum(temp_result * temp_result)
        n = temp_result.shape[0]
        C = (E2/n - E*E/n/n)/T/T*100
        Cs.append(C)
    sns.lineplot(Ts,Cs)
    sns.scatterplot(Ts,Cs)


plt.legend(['N=11','N=21','N=51','N=101','N=121','N=151'])
ax.set_xlabel("$T$")
ax.set_ylabel("$C(T)$")
ax.set_title("Ising - Python")
plt.show()
```

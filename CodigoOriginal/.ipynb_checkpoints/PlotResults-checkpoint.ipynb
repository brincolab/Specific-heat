```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
sns.set()
%matplotlib inline
```

```python
fig, ax = plt.subplots(1,1,figsize=(10,5))

file_lists = ["Res2N11", "Res2N21", "Res2N51", "Res2N101", "Res2N121", "Res2N151"]
for file in file_lists:
    x = []
    y = []
    y2 =[]
    
    f = open(file)
    for line in f.readlines():
        result = [el for el in line.replace("\n","").split(" ") if el]
        x.append(result[0])
        y.append(result[1])
        y2.append(result[2])
    f.close()
    
    #String to Float
    x = np.array(x,dtype=np.float32)
    y = np.array(y,dtype=np.float32)
    y2 = np.array(y2,dtype=np.float32)
    
    sns.lineplot(x,y2)
    sns.scatterplot(x,y2)


plt.legend(['N=11','N=21','N=51','N=101','N=121','N=151'])
ax.set_xlabel("$T$")
ax.set_ylabel("$C(T)$")
ax.set_title("Ising - Fortran")
plt.show()
```
